```python
#Finding categorical features
num_cols = df._get_numeric_data().columns

cate_cols = list(set(df.columns)-set(num_cols))
cate_cols.remove('Cat')
# cate_cols.remove('Attack Type')

cate_cols
```

```python
df.drop('Timestamp',axis = 1,inplace = True)
df.drop('Src_IP',axis = 1,inplace = True)
df.drop('Dst_IP',axis = 1,inplace = True)
df.drop('Sub_Cat',axis = 1,inplace = True)
df.drop('Label',axis = 1,inplace = True)
df.drop('Flow_ID',axis = 1,inplace = True)
```

```python
mydf = df.copy()
pmap = {'DoS':1,'Normal':0}
mydf['Cat'] = mydf['Cat'].map(pmap)
mydf.head(99464)
```

```python
def clean_dataset(df):
    assert isinstance(df, pd.DataFrame), "df needs to be a pd.DataFrame"
    df.dropna(inplace=True)
    indices_to_keep = ~df.isin([np.nan, np.inf, -np.inf]).any(1)
    return df[indices_to_keep].astype(np.float64)

mydf = clean_dataset(mydf)
```

```python

from genetic_selection import GeneticSelectionCV
# from sklearn_genetic import GASearchCV

from sklearn.tree import DecisionTreeClassifier
import pandas as pd
import numpy as np
np.random.seed(42) 


X = mydf.drop(['Cat'], axis=1)
y = mydf[['Cat']].astype(float)
estimator = DecisionTreeClassifier()
model = GeneticSelectionCV(
    estimator, cv=5, verbose=0,
    scoring="accuracy", max_features=20,
    n_population=100, crossover_proba=0.5,
    mutation_proba=0.2, n_generations=50,
    crossover_independent_proba=0.5,
    mutation_independent_proba=0.04,
    tournament_size=3, n_gen_no_change=10,
    caching=True, n_jobs=-1)
model = model.fit(X, y)

# print the important features using the generic algorithm
print('Features:', X.columns[model.support_],(X.columns[model.support_]).dtype)
```

```python
print(X.columns[model.support_])
yii=X.columns[model.support_]
tt=yii.tolist()
print(tt)
```

```python
#Decision Tree
from sklearn.tree import DecisionTreeClassifier
import time
```

```python
model2 = DecisionTreeClassifier()
```

```python
x_train_GA=mydf[tt]
y_train_ga = mydf[['Cat']].astype(float)

x_test_GA=mydf[tt]
y_test_ga = mydf[['Cat']].astype(float)
```

```python
# print(Y_test)
arr = y_test_ga.to_numpy()
import numpy as np
result_tot_ga = arr.flatten()
# result_tot
```

```python
start_time = time.time()
model2.fit(x_train_GA, y_train_ga.values.ravel())
end_time = time.time()
print("Trining time: ",end_time-start_time)
print("Train score is:", model2.score(x_train_GA, y_train_ga))
```

```python
start_time = time.time()
Y_pred_GA = model2.predict(x_test_GA)
end_time = time.time()
print("Testing time: ",end_time-start_time)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
cm = confusion_matrix(Y_pred_GA, result_tot_ga)
ConfusionMatrixDisplay(cm).plot()
print(confusion_matrix(Y_pred_GA, result_tot_ga))
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
# accuracy: (tp + tn) / (p + n)
accuracy = accuracy_score(y_test_ga, Y_pred_GA)
print('Accuracy: %f' % accuracy)
# precision tp / (tp + fp)
precision = precision_score(y_test_ga, Y_pred_GA)
print('Precision: %f' % precision)
# recall: tp / (tp + fn)
recall = recall_score(y_test_ga, Y_pred_GA)
print('Recall: %f' % recall)
# f1: 2 tp / (2 tp + fp + fn)
f1 = f1_score(y_test_ga, Y_pred_GA)
print('F1 score: %f' % f1)
```
